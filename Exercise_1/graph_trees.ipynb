```python
import matplotlib.pyplot as plt
import networkx as nx

k = 3
d = 3

T = nx.balanced_tree(k, d)

pos = nx.nx_agraph.graphviz_layout(T, prog="dot", args="-Eminlen=3")

options = {
    "font_size": 10,
    "node_size": 300,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 1,
    "width": 1,
    "with_labels": True
}

plt.figure(figsize=(10, 5))
nx.draw(T, pos, **options)
plt.margins(0)
plt.axis("equal")
# plt.savefig("tree_a.pdf")
plt.show()
```
