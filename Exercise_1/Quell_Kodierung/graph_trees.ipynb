```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

save_figures_to_latex = False
figures_path = '/home/tellocam/Projects/HPC/Exercise_1/latex_ex_1/figures'
```

```python
k_a = 3
d_a = 3

T = nx.balanced_tree(k_a, d_a)

pos = nx.nx_agraph.graphviz_layout(T, prog="dot", args="-Eminlen=3")

options = {
    "font_size": 13,
    "node_size": 500,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 1,
    "width": 2,
    "with_labels": True
}

plt.figure(1)
plt.figure(figsize=(15, 5))
nx.draw(T, pos, **options)
plt.margins(0)
plt.axis("equal")

if (save_figures_to_latex == True):
    plt.savefig(figures_path + '/Ex1_2_a.pdf')
```

```python
k_b = 3
d_b = 4

T = nx.Graph()

counter = 0
for i in range(d_b):
    for j in range(k_b - 1):
        for n in range(k_b ** i):
            counter += 1
            T.add_edge(n, counter)

pos = nx.nx_agraph.graphviz_layout(T, prog="dot", args="-Eminlen=3")

options = {
    "font_size": 13,
    "node_size": 500,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 1,
    "width": 2,
    "with_labels": True
}

plt.figure(2)
plt.figure(figsize=(20, 5))
nx.draw(T, pos, **options)
plt.margins(0)
plt.axis("equal")

if (save_figures_to_latex == True):
    plt.savefig(figures_path + '/Ex1_2_b.pdf')
```

```python
max_nodes = 9

T = nx.Graph()
for i in range(1, max_nodes):
    if i % 2 == 0:
        T.add_edge(i - 1, i + 1)
    else:
        T.add_edge(i, i + 1)

pos = nx.nx_agraph.graphviz_layout(T, prog="dot", args="")
options = {
    "font_size": 13,
    "node_size": 500,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 1,
    "width": 2,
    "with_labels": True
}

plt.figure(3)
plt.figure(figsize=(5, 5))
nx.draw(T, pos, **options)
plt.margins(0)
plt.axis("equal")

if (save_figures_to_latex == True):
    plt.savefig(figures_path + '/Ex1_2_c.pdf')
```
